```python
import pandas as pd

# Convert YEAR, MONTH, DAY_WEEK, and HOUR to string for concatenation
df['YEAR'] = df['YEAR'].astype(str)
df['MONTH'] = df['MONTH'].astype(str)
df['DAY_WEEK'] = df['DAY_WEEK'].astype(str)
df['HOUR'] = df['HOUR'].astype(str)

# Ensure that DAY_WEEK and HOUR have valid values (1-7 for days, 0-23 for hours)
valid_hours = df['HOUR'].astype(int).between(0, 23)
valid_days = df['DAY_WEEK'].astype(int).between(1, 7)

# Filter out invalid rows
df = df[valid_hours & valid_days]

# Combine year, month, day, and hour to create a datetime column
df['DATETIME'] = pd.to_datetime(df['YEAR'] + '-' + df['MONTH'] + '-' + df['DAY_WEEK'] + ' ' + df['HOUR'], format='%Y-%m-%d %H')

# Handle missing values
df.fillna({'ALCOHOLNAME': 'Unknown', 'WEATHER1NAME': 'Unknown', 'LGT_CONDNAME': 'Unknown'}, inplace=True)

# Extract day of week and hour for analysis
df['DAY_OF_WEEK'] = df['DATETIME'].dt.day_name()
df['HOUR_OF_DAY'] = df['DATETIME'].dt.hour

# Now we can proceed with the analysis
```

```python
# Summary statistics
print(df.describe())

# Group by weather conditions and count accidents
weather_accidents = df.groupby('WEATHER1NAME').size().sort_values(ascending=False)
print(weather_accidents)

# Group by lighting conditions and count accidents
lighting_accidents = df.groupby('LGT_CONDNAME').size().sort_values(ascending=False)
print(lighting_accidents)

# Group by hour of day and count accidents
hourly_accidents = df.groupby('HOUR_OF_DAY').size().sort_values(ascending=False)
print(hourly_accidents)

# Group by day of week and count accidents
weekly_accidents = df.groupby('DAY_OF_WEEK').size().sort_values(ascending=False)
print(weekly_accidents)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the matplotlib figure
plt.figure(figsize=(12, 8))

# Plot accidents by weather conditions
plt.subplot(2, 2, 1)
weather_accidents.plot(kind='bar')
plt.title('Accidents by Weather Conditions')
plt.ylabel('Number of Accidents')

# Plot accidents by lighting conditions
plt.subplot(2, 2, 2)
lighting_accidents.plot(kind='bar')
plt.title('Accidents by Lighting Conditions')
plt.ylabel('Number of Accidents')

# Plot accidents by hour of day
plt.subplot(2, 2, 3)
hourly_accidents.plot(kind='bar')
plt.title('Accidents by Hour of Day')
plt.ylabel('Number of Accidents')

# Plot accidents by day of week
plt.subplot(2, 2, 4)
weekly_accidents.plot(kind='bar')
plt.title('Accidents by Day of Week')
plt.ylabel('Number of Accidents')

plt.tight_layout()
plt.show()
```

```python
# Split the data into two groups based on alcohol involvement
df_with_alcohol = df[df['ALCOHOLNAME'] == 'Alcohol Involved']
df_without_alcohol = df[df['ALCOHOLNAME'] == 'No Alcohol Involved']
```

```python
import matplotlib.pyplot as plt

# Group by weather conditions and count accidents
weather_accidents_with_alcohol = df_with_alcohol.groupby('WEATHER1NAME').size().sort_values(ascending=False)
weather_accidents_without_alcohol = df_without_alcohol.groupby('WEATHER1NAME').size().sort_values(ascending=False)

# Set up the matplotlib figure
plt.figure(figsize=(14, 6))

# Plot accidents by weather conditions (with alcohol)
plt.subplot(1, 2, 1)
weather_accidents_with_alcohol.plot(kind='bar', color='red')
plt.title('Accidents by Weather Conditions (With Alcohol)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

# Plot accidents by weather conditions (without alcohol)
plt.subplot(1, 2, 2)
weather_accidents_without_alcohol.plot(kind='bar', color='blue')
plt.title('Accidents by Weather Conditions (Without Alcohol)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Group by lighting conditions and count accidents
lighting_accidents_with_alcohol = df_with_alcohol.groupby('LGT_CONDNAME').size().sort_values(ascending=False)
lighting_accidents_without_alcohol = df_without_alcohol.groupby('LGT_CONDNAME').size().sort_values(ascending=False)

# Set up the matplotlib figure
plt.figure(figsize=(14, 6))

# Plot accidents by lighting conditions (with alcohol)
plt.subplot(1, 2, 1)
lighting_accidents_with_alcohol.plot(kind='bar', color='red')
plt.title('Accidents by Lighting Conditions (With Alcohol)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

# Plot accidents by lighting conditions (without alcohol)
plt.subplot(1, 2, 2)
lighting_accidents_without_alcohol.plot(kind='bar', color='blue')
plt.title('Accidents by Lighting Conditions (Without Alcohol)')
plt.ylabel('Number of Accidents')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```
